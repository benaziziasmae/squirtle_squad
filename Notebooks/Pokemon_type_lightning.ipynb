```python
# import dependencies
import pandas as pd
import requests
import csv
```

```python
# Call API
url = 'https://api.pokemontcg.io/v2/types'
```

```python
card_info = requests.get(url)
card_info
```

```python
card_info.json()
```

```python
# Fiter cards by type
type_url= 'https://api.pokemontcg.io/v2/cards?q=supertype:pokemon types:Lightning'
type_info = requests.get(type_url)
type_info
```

```python
type_info.json()
```

```python
lightning_df = pd.DataFrame(type_info.json()['data'])
lightning_df
```

```python
lightning_df.columns
```

```python
lightning_sample = lightning_df.head(10)
lightning_sample
```

```python
lightning_sample.to_csv('sample/type_lightning_sample.csv', columns=lightning_sample.columns)
```
