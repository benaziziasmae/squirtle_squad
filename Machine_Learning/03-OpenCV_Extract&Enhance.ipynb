```python
import os
import numpy as np
import sys
import cv2
import matplotlib.pyplot as plt

try:
    from PIL import Image
except ImportError:
    import Image
```

```python
file_dir = os.path.abspath('')+"\\train\\fullcard"
file_path = os.path.join(file_dir, "corviknight_vmax.png")
image_file = cv2.imread(file_path)

# convert to grayscale
gray_image = cv2.cvtColor(image_file,cv2.COLOR_BGR2GRAY)

# write file as gray
cv2.imwrite(f'{file_dir}/corviknight_vmax_gray.png', gray_image)
```

```python
cv2.imshow('image', image_file)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
