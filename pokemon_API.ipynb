```python
#Import Dependencies
import pandas as pd
import json
import requests
from config import api_key
```

### Pokemon Sets

```python
#Query URL
query_sets = "https://api.pokemontcg.io/v2/sets"
```

```python
#Getting Sets
get_pokemon_sets = requests.get(query_sets)
pokemonsets_json = get_pokemon_sets.json()
print(pokemonsets_json)
```

```python
pokemon_sets_df = pd.DataFrame(get_pokemon_sets.json()["data"])
pokemon_sets_df
```

```python
pokemon_sets_df.columns
```

```python
pokemon_sets_df.to_csv("datasets/pokemon_sets.csv", columns=pokemon_sets_df.columns)
```

### Pokemon Subtypes

```python
#Query URL
query_subtypes = "https://api.pokemontcg.io/v2/subtypes"
```

```python
#Getting Subtypes
get_pokemon_subtypes = requests.get(query_subtypes)
pokemonsubtypes_json = get_pokemon_subtypes.json()
print(pokemonsubtypes_json)
```

```python
#Filtering cards by Subtypes
query_baby = "https://api.pokemontcg.io/v2/cards?q=supertype:pokemon subtypes:baby"
subtype_baby = requests.get(query_baby)
subtypebaby_json = subtype_baby.json()
print(subtypebaby_json)
```

```python
babypokemon_df = pd.DataFrame(subtype_baby.json()["data"])
babypokemon_df
```

```python
babypokemon_df.columns
```

```python
babypokemon_df.to_csv("datasets/babypokemon.csv", columns=babypokemon_df.columns)
```
